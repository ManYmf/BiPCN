```python
from radas import run_experiment, get_ray_results_dir
import ray.tune as tune
from v92 import trainable
import torch

results = await run_experiment(
    user_name="mengfan",
    experiment_name="å¤ç°è®ºæ–‡",
    param_space=dict(
    epochs        = tune.grid_search([50]),
    batch_size    = tune.grid_search([256]),

    hidden_dims   = tune.grid_search([[128, 128]]),
    activation   = tune.grid_search(["tanh","relu"]),

    steps_train   = tune.grid_search([10]),
    steps_test    = tune.grid_search([8]),
    alpha_disc   = tune.grid_search([1.0]),

    lr_x          = tune.grid_search([0.02]),
    momentum_x   = tune.grid_search([0.5]),

    param_optim  = tune.grid_search(["adamw"]),
    lr_theta     = tune.grid_search([1e-3]),
    weight_decay = tune.grid_search([1e-4]),

    seed          = tune.grid_search([0, 1, 2]),

    ),
    # IMPORTANT: when using cluster, your trainable cannot be defined in this script, but has to be defined in a separate file, and imported in this script
    trainable=trainable,
    # When use cluster, local_storage_path needs to be specified in this way
    local_storage_path=get_ray_results_dir(),
    # Specify the cluster to use
    run_with="cluster:atol-gpu-5090",
    # Specify gram per trial
    cpu_per_trial=1,
    # If the gram_per_trial you set is larger than GRAM of a single GPU, you will be allocated with more than one GPUs
    # to fit your need. But in this case, you will need to handle how do you want to distribute your workload to these multiple GPUs
    # To check GRAM of a single GPU, use the following code:
    # from radas.clusters import gram_per_gpu
    #
    runtime_env={
        "pip": ["emoji"],
        # for more runtime_env specifications, see https://docs.ray.io/en/latest/ray-core/handling-dependencies.html#api-reference
    },
    #
    instant_gpu_allocation=1,
)
```

```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------
# 1) æŠŠ wide-format çš„ epoch åˆ—æŠ½å‡ºæ¥ï¼Œè½¬æˆ long-format
# -------------------------
def make_long_curves(df, metric="test_acc"):
    # æ‰¾åˆ°å½¢å¦‚ epoch_12/test_acc çš„åˆ—
    pat = re.compile(rf"^epoch_(\d+)/{re.escape(metric)}$")
    cols = []
    for c in df.columns:
        m = pat.match(c)
        if m:
            cols.append((int(m.group(1)), c))
    cols = sorted(cols, key=lambda x: x[0])
    if not cols:
        raise ValueError(f"No columns matched metric={metric}")

    epochs = [e for e, _ in cols]
    metric_cols = [c for _, c in cols]

    # ç”¨ melt è½¬ long
    base_cols = [c for c in df.columns if not c.startswith("epoch_")]
    tmp = df[base_cols + metric_cols].copy()
    long = tmp.melt(
        id_vars=base_cols,
        value_vars=metric_cols,
        var_name="epoch_col",
        value_name=metric
    )
    long["epoch"] = long["epoch_col"].str.extract(r"epoch_(\d+)/")[0].astype(int)
    long = long.drop(columns=["epoch_col"]).sort_values(["config/seed", "epoch"], kind="stable")
    return long, epochs

long_acc, epochs = make_long_curves(df, metric="test_acc")

# ç»™æ¯ä¸ª trial ä¸€ä¸ªç®€æ´æ ‡ç­¾ï¼ˆä½ ä¹Ÿå¯æ¢æˆ trial_idï¼‰
def trial_label(row):
    return f"{row['config/activation']}-seed{row['config/seed']}"

df = results["df"].copy()
df["trial_label"] = df.apply(trial_label, axis=1)

# -------------------------
# 2) å›¾1ï¼šæ¯ä¸ª trial çš„å­¦ä¹ æ›²çº¿ï¼ˆtest_acc vs epochï¼‰
# -------------------------
plt.figure()
for i, row in df.iterrows():
    y = [row.get(f"epoch_{e}/test_acc", np.nan) for e in epochs]
    plt.plot(epochs, y, label=row["trial_label"])
plt.xlabel("Epoch")
plt.ylabel("Test Accuracy")
plt.title("Learning Curves (test_acc)")
plt.legend(fontsize=8, ncol=2)
plt.tight_layout()
plt.show()

# -------------------------
# 3) å›¾2ï¼šæ¯ä¸ª trial çš„â€œæœ€ä½³ç²¾åº¦ / æœ€ç»ˆç²¾åº¦â€å¯¹æ¯”
# -------------------------
def best_and_last(df, epochs):
    best = []
    last = []
    for _, row in df.iterrows():
        y = np.array([row.get(f"epoch_{e}/test_acc", np.nan) for e in epochs], dtype=float)
        best.append(np.nanmax(y))
        last.append(y[-1])
    out = df[["trial_label", "config/activation", "config/seed", "config/lr_theta", "config/lr_x", "config/momentum_x"]].copy()
    out["best_acc"] = best
    out["last_acc"] = last
    return out

summ = best_and_last(df, epochs)

plt.figure()
x = np.arange(len(summ))
plt.bar(x - 0.2, summ["best_acc"], width=0.4, label="best_acc")
plt.bar(x + 0.2, summ["last_acc"], width=0.4, label="last_acc")
plt.xticks(x, summ["trial_label"], rotation=45, ha="right")
plt.ylabel("Accuracy")
plt.title("Best vs Last Test Accuracy per Trial")
plt.legend()
plt.tight_layout()
plt.show()

# -------------------------
# 4) å›¾3ï¼šæŒ‰ activation åˆ†ç»„çš„ best_accï¼ˆå‡å€¼Â±æ ‡å‡†å·®ï¼‰
# -------------------------
g = summ.groupby("config/activation")["best_acc"]
means = g.mean()
stds = g.std(ddof=0).fillna(0.0)

plt.figure()
x = np.arange(len(means))
plt.errorbar(x, means.values, yerr=stds.values, fmt="o")
plt.xticks(x, means.index)
plt.ylabel("Best Test Accuracy")
plt.title("Best Accuracy by Activation (mean Â± std)")
plt.tight_layout()
plt.show()

# -------------------------
# 5) å›¾4ï¼šlr_theta vs best_accï¼ˆå»ºè®®å¯¹ lr å– log è½´ï¼‰
# -------------------------
plt.figure()
plt.scatter(summ["config/lr_theta"], summ["best_acc"])
plt.xscale("log")
plt.xlabel("lr_theta (log scale)")
plt.ylabel("best_acc")
plt.title("Hyperparam Sensitivity: lr_theta vs best_acc")
plt.tight_layout()
plt.show()

# -------------------------
# 6) å¯é€‰ï¼šæŠŠ E_test_after_infer ä¹Ÿç”»å‡ºæ¥ï¼ˆä½†ä½ ç°åœ¨å®ƒå…¨æ˜¯ 0ï¼Œæ‰€ä»¥ä¼šæ˜¯ä¸€æ¡ç›´çº¿ï¼‰
# -------------------------
if any(c.endswith("/E_test_after_infer") for c in df.columns):
    long_E, epochs_E = make_long_curves(df, metric="E_test_after_infer")
    plt.figure()
    for i, row in df.iterrows():
        y = [row.get(f"epoch_{e}/E_test_after_infer", np.nan) for e in epochs_E]
        plt.plot(epochs_E, y, label=row["trial_label"])
    plt.xlabel("Epoch")
    plt.ylabel("E_test_after_infer")
    plt.title("Learning Curves (E_test_after_infer)")
    plt.legend(fontsize=8, ncol=2)
    plt.tight_layout()
    plt.show()
```
